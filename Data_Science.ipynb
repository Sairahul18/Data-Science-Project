```python
#grouped bar graph
import matplotlib.pyplot as plt
import numpy as np
#defining categories and their corresponding values
categories=['Lion','Horse','Kangaroo']
values1=[15,24,30]
values2=[20,18,25]
#Setting the width of the bars
bar_width=0.35
#calculating bar positions for both groups
bar_positions1=np.arange(len(categories))
bar_positions2=bar_positions1+bar_width
#Creating the first set of bars(Group1)
plt.bar(bar_positions1,values1,width=bar_width,label='2021',color='skyblue')
#Create the second set of bars(Group 2)next to the
plt.bar(bar_positions2,values2,width=bar_width,label='2022',color='orange')
#adding labels to the axes
plt.xlabel('Categories')
plt.ylabel('Values')
#Adding a title to the graph
plt.title('Grouped Bar Graph')
#Displaying a legend to identify the groups
plt.legend()
#showing the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
#Defining categories and values for two groups
categories=['Transport','Income Tax','Housing']
values1=[15,24,30]
values2=[20,18,25]
#Creating the first set of bars(Group 1)without any of set
plt.bar(categories,values1,label='First Half',color='skyblue')
#creating the second set of bars(Group2)plotted with'bottom' set to the values
#This make Group 2 bars stacked on top of Group 1 bars
plt.bar(categories,values2,bottom=values1,label='second Half',color='orange')
#Adding labels to the axes
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Graph')
plt.legend()
plt.show()
```

```python
import pandas as pd
#Series generated from a list
a=[1,7,2]
myvar =pd.Series(a)
print(a)
print(myvar)
print('------------------')
#labels
print(myvar[0])
print('------------------')
#Assigning seperate labels to values in the series
myvar2=pd.Series(a,index=["x","y","z"])
print(myvar2)
print(myvar2["y"])
print('-------------------')
#Series generated from dictionary
calories={"day1":420,"day2":380,"day3":390}
myvar=pd.Series(calories)
print(myvar)
print('---------------------')
myvar=pd.Series(calories,index=["day1","day2"])
print(myvar)
print('----------------------')
#Series generated from Scalar Value
series=pd.Series(5,index=[0,1,2,3])
print(series)
print('-----------------------')
```

```python
#Integer Series
int_series=pd.Series([1,2,3,4])
print(int_series)
print('--------------')
#Float Series
float_series=pd.Series([1.1,2.2,3.3,4.4])
print(float_series)
print('----------------')
#String Series
string_series=pd.Series(['a','b','c','d'])
print(string_series)
print('-------------------')
#Label-based indexing
data={'a':10,'b':20,'c':30}
series=pd.Series(data)
print(series['a'])
print(series[['a','c']])
print('-------------------')
#boolean indexing
data=[10,20,30,40,50]
series=pd.Series(data)
#Access elements greater than 25
print(series[series>25])
print('-------------------')

```

```python
#Creating  a data frame
data={
    'Name':['Alice','Bob','Charlie'],
    'Age':[25,30,35],
    'City':['New York','Los Angeles','Chicago']
}
df=pd.DataFrame(data)
print('----------------------------')
#Selecting Columns
print(df.Name)
print(df['Name'])
print(df[['Name','City']])
print('-----------------------------')
#select a single row by label
print(df.loc[0])
print(df.iloc[0])
#select multiple rows
print(df.iloc[0:2])
print('------------------------')
#Conditional Selection
data={
    'Name':['Alice','Bob','Charlie'],
    'Age':[25,30,35],
    'City':['New York','Los Angeles','Chicago']
}
df=pd.DataFrame(data)
#Select rows where Age is greater than 25
print(df[df['Age']>25])
print('-------------------------------')
print(df.loc[(df.Name=='Bob')&(df.City=='Los Angeles')])
```

```python
#conctenate a mix of series and dataframe
#importing pandas module
import pandas as pd
#Define a dictionary containing employee data
data1={'Name':['Jai','Princi','Gaurav','Anuj'],
       'Age':[27,24,22,32],
       'Address':['Nagpur','Kanpur','Allahabad','kannuaj'],
       'Qualification':['Msc','MA','MCA','Phd']}

#convert the dictionary into DataFrame
df=pd.DataFrame(data1,index=[0,1,2,3])
#creating a series
s1=pd.Series([1000,2000,3000,4000],name='Salary')
print(df,"\n\n",s1)
#combining series and dataframe
res=pd.concat([df,s1],axis=1)
res
```
